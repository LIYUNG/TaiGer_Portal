## Web scrap with BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup; # print the Html in the webpage
```

```python
table = soup.find_all('table')[1] # this is manually check which tabel do we want
```

```python
world_titles = table.find_all('th') # this is the secondary title
```

```python
print(table)
```

```python
world_titles # still in html format
```

```python
world_table_titles = [title.text.strip() for title in world_titles] # extract the "text" from html format
```

```python
world_table_titles
```

## Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr') # r stands for row
```

```python
for count, row in enumerate(column_data[1:]):
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data] # we want to append each output into the dataframe
    
    df.loc[count] = individual_row_data
```
